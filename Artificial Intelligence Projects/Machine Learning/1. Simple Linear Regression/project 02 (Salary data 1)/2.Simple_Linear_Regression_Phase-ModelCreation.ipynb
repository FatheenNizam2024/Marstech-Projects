```python
#importing Libraries
import numpy as np   # using for arithmetic calculation
import matplotlib.pyplot as plt   # using to plot a graph 
import pandas as pd    # this use for handling table method data sets library
```

```python
dataset = pd.read_csv("Salary_Data.csv") 
```

```python
dataset
```

```python
independent = dataset[["Years of Experience"]]
```

```python
independent
```

```python
dependent = dataset[["Salary"]]
```

```python
dependent
```

```python
# using library from sklearn (file) model_selection (class) and import from train_test_split (Function)
# this use for split our data for trainning and testing 
from sklearn.model_selection import train_test_split

# here "train_test_split" work our data split 4 types (input data train),(input data test),(output data train),(output data test)
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.20, random_state=0)
```

```python
# create model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)  # model , weight , bias also in  regressor 
```

```python
# finding weight from regressor
weight = regressor.coef_
weight
```

```python
# finding bias from regressor
bias = regressor.intercept_
bias
```

```python
# Evaluated 
# finding Predict output value for the corresponding Actual value
y_pred = regressor.predict(X_test)   # X_test - is the Actual Value
```

```python
# finding our 
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)  # r2_score ("Actual", "Predict")
```

```python
r_score
```
